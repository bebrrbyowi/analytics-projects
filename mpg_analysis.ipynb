**Задание 5.5**  

Используя цикл while и счетчик, найдите первый год выпуска (model_year), в котором средний расход топлива превысил 25 миль на галлон. Начните проверку с 1970 года (model_year = 70).

Подсказка: увеличивайте счетчик на 1 и проверяйте условие для каждого года.

````python
import pandas as pd # Импортируем библиотеку pandas для работы с табличными данными (DataFrame)
import seaborn as sns # Импортируем библиотеку seaborn для загрузки демонстрационных наборов данных

# Проверяем, существует ли DataFrame 'mpg'. Если нет, загружаем его из seaborn.
# Это нужно, чтобы код работал, даже если предыдущая ячейка с загрузкой 'mpg' не была выполнена.
if 'mpg' not in locals():
    mpg = sns.load_dataset("mpg")

current_year = 70 # Инициализируем текущий год, начиная с 1970 года (представленного как 70)
found_year = None # Переменная для хранения первого найденного года

# Запускаем цикл while, который будет продолжаться до тех пор, пока не будет найден год
# или пока мы не проверим все года в диапазоне model_year (до 82).
# В данном случае, мы будем искать до 82 (1982 год), что является максимальным в датасете.
while current_year <= mpg['model_year'].max():
    # Фильтруем данные для текущего года
    year_data = mpg[mpg['model_year'] == current_year]

    # Проверяем, есть ли данные за этот год, чтобы избежать ошибок деления на ноль
    if not year_data.empty:
        # Рассчитываем средний расход топлива для текущего года
        average_mpg_for_year = year_data['mpg'].mean()

        # Проверяем, превышает ли средний расход топлива 25 миль на галлон
        if average_mpg_for_year > 25:
            found_year = 1900 + current_year # Преобразуем сокращенный год в полный формат (например, 70 -> 1970)
            break # Если условие выполнено, прерываем цикл

    current_year += 1 # Увеличиваем счетчик года на 1 для перехода к следующему году

# Выводим результат
if found_year:
    print(f"Первый год выпуска, в котором средний расход топлива превысил 25 миль на галлон: {found_year}")
else:
    print("Год, в котором средний расход топлива превысил 25 миль на галлон, не найден в данных.")
````

**Задание 5.8**  

Напишите функцию analyze_category(df, category_column, numeric_column, threshold), которая:

* Принимает DataFrame, название категориального столбца, название числового столбца и пороговое значение
* Для каждой категории определяет, превышает ли среднее значение числового столбца указанный порог
* Возвращает словарь, где ключи — категории, а значения — True/False в зависимости от выполнения условия

Проверьте, для каких стран производства средний расход топлива превышает 25 миль на галлон.

Вызовите функцию: analyze_category(mpg, 'origin', 'mpg', 25) и выведите результат.

````python
import pandas as pd # Импортируем библиотеку pandas для работы с табличными данными (DataFrame)
import seaborn as sns # Импортируем библиотеку seaborn для загрузки демонстрационных наборов данных

# Проверяем, существует ли DataFrame 'mpg'. Если нет, загружаем его из seaborn.
# Это нужно, чтобы код работал, даже если предыдущая ячейка с загрузкой 'mpg' не была выполнена.
if 'mpg' not in locals():
    mpg = sns.load_dataset("mpg")

def analyze_category(df, category_column, numeric_column, threshold): # Определяем функцию analyze_category
    result = {} # Инициализируем пустой словарь для хранения результатов
    # Группируем DataFrame по категориальному столбцу и рассчитываем среднее для числового столбца
    grouped_data = df.groupby(category_column)[numeric_column].mean()

    # Итерируемся по каждой категории и ее среднему значению
    for category, mean_value in grouped_data.items():
        # Проверяем, превышает ли среднее значение пороговое
        result[category] = mean_value > threshold
    return result # Возвращаем словарь с результатами

# Вызываем функцию analyze_category для DataFrame mpg:
# category_column = 'origin' (страна производства)
# numeric_column = 'mpg' (расход топлива)
# threshold = 25 (пороговое значение расхода топлива)
analysis_result = analyze_category(mpg, 'origin', 'mpg', 25)

# Выводим полученный результат
print(analysis_result)
````

````python
import pandas as pd # Импортируем библиотеку pandas для работы с табличными данными (DataFrame)
import seaborn as sns # Импортируем библиотеку seaborn для загрузки демонстрационных наборов данных

# Проверяем, существует ли DataFrame 'mpg'. Если нет, загружаем его из seaborn.
# Это нужно, чтобы код работал, даже если предыдущая ячейка с загрузкой 'mpg' не была выполнена.
if 'mpg' not in locals():
    mpg = sns.load_dataset("mpg")

def find_outliers(df, column, multiplier): # Определяем функцию для поиска выбросов
    mean = df[column].mean() # Рассчитываем среднее значение для указанного столбца
    std = df[column].std() # Рассчитываем стандартное отклонение для указанного столбца
    # Находим выбросы: значения, которые отклоняются от среднего более чем на (multiplier * std)
    outliers = df[abs(df[column] - mean) > multiplier * std]
    return outliers # Возвращаем DataFrame, содержащий только строки с выбросами

# Вызываем функцию find_outliers для столбца 'mpg' с множителем 2
outliers_mpg = find_outliers(mpg, 'mpg', 2)

# Выводим количество найденных выбросов
print(f"Количество найденных выбросов в столбце '{'mpg'}' (multiplier=2): {len(outliers_mpg)}")

# Выводим первые 5 записей с выбросами (название автомобиля, расход топлива, страна)
print("\nПервые 5 записей с выбросами:")
print(outliers_mpg[['name', 'mpg', 'origin']].head())
````
